```python
model = tf.keras.models.load_model('Heart_CNN.model')
```

```python
data = pd.read_csv('data/Data_Entry_2017.csv')
test_list = pd.read_csv('data/test_list.txt', sep="\n", header = None)
```

```python
test = test_list[0].tolist()
test_data = data[data['Image Index'].isin(test)]
test_data[test_data['Finding Labels'].str.contains('Cardiomegaly')]
```

```python
def prepare_img(file):
    IMG_SIZE = 512
    dim = (IMG_SIZE, IMG_SIZE)
    img_array = cv2.imread(file, cv2.IMREAD_GRAYSCALE)/255.0
    new_array = cv2.resize(img_array, dim)
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)

test_image1 = 'data/images/images_001/images/00000013_025.png'
test_image2 = 'data/images/images_012/images/00030412_002.png'
test_image3 = 'data/images/images_012/images/00030535_000.png'
test_image4 = 'data/images/images_012/images/00030535_001.png'
test_image5 = 'data/images/images_012/images/00030608_000.png'
test_image6 = 'data/images/images_012/images/00030608_001.png'

prediction1 = model.predict([prepare_img(test_image1)])
prediction2 = model.predict([prepare_img(test_image2)])
prediction3 = model.predict([prepare_img(test_image3)])
prediction4 = model.predict([prepare_img(test_image4)])
prediction5 = model.predict([prepare_img(test_image5)])
prediction6 = model.predict([prepare_img(test_image6)])
```

```python
ls = [prediction1[0][0],prediction2[0][0], prediction3[0][0], prediction4[0][0], prediction5[0][0], prediction6[0][0]]
ind = ['00000013_025.png','00030412_002.png','00030535_000.png','00030535_001.png','00030608_000.png','00030608_001.png']
labels = ['Cardiomegaly|Infiltration|Mass|Nodule','Cardiomegaly|Effusion','Cardiomegaly','Cardiomegaly','Cardiomegaly','Cardiomegaly']
pred = pd.DataFrame({'values': ls, 'actual label': labels}, index = ind)
pred
```
